## Introduction

In this project, we will be cleaning and exploring a Star Wars survey data set. This survey was conducted by FiveThirtyEight -- they surveyed Star Wars fans using the online tool SurveyMonkey. They received 835 total responses.

## Reading the data set

```python
import pandas as pd
star_wars = pd.read_csv('star_wars.csv', encoding="ISO-8859-1")
```

```python
# Investigating what columns there are
star_wars.columns
```

```python
star_wars.head(10)
```

There are several other columns containing answers to questions about the Star Wars movies. For some questions, the respondent had to check one or more boxes. This type of data is difficult to represent in columnar format. As a result, this data set needs a lot of cleaning.

## Cleaning the data

```python
# Removing rows where RespondentID is NaN
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
# For the next two columns, we convert the Yes/No responses to True/False
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

```python
# Creating a dictionary to define a mapping
yes_no = {
    "Yes": True,
    "No": False
}

# Converting the two columns
for col in [
    'Have you seen any of the 6 films in the Star Wars franchise?'
    , 'Do you consider yourself to be a fan of the Star Wars film franchise?'
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

```python
# Converting columns 3-8 to True/False

import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
        star_wars[col] = star_wars[col].map(movie_mapping)

star_wars.head()
```

```python
# Renaming the column names

new_col_names = {
    'Which of the following Star Wars films have you seen? Please select all that apply.': "seen_1"
    , 'Unnamed: 4': "seen_2"
    , 'Unnamed: 5': "seen_3"
    , 'Unnamed: 6': "seen_4"
    , 'Unnamed: 7': "seen_5"
    , 'Unnamed: 8': "seen_6"

}

star_wars = star_wars.rename(columns=new_col_names)
star_wars.columns
```

```python
star_wars.head()
```

The next six columns ask the respondent to rank the Star Wars movies in order of least favorite to most favorite. 1 means the film was the most favorite, and 6 means it was the least favorite. Each of the following columns can contain the value 1, 2, 3, 4, 5, 6, or NaN:

- Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film. - How much the respondent liked Star Wars: Episode I The Phantom Menace
- Unnamed: 10 - How much the respondent liked Star Wars: Episode II Attack of the Clones
- Unnamed: 11 - How much the respondent liked Star Wars: Episode III Revenge of the Sith
- Unnamed: 12 - How much the respondent liked Star Wars: Episode IV A New Hope
- Unnamed: 13 - How much the respondent liked Star Wars: Episode V The Empire Strikes Back
- Unnamed: 14 - How much the respondent liked Star Wars: Episode VI Return of the Jedi

Fortunately, these columns don't require a lot of cleanup. We'll need to convert each column to a numeric type, though, then rename the columns so that we can tell what they represent more easily.

```python
# Converting the columns to a numeric type
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# Renaming the columns with a more descriptive name
new_col_names = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': "ranking_1"
    , 'Unnamed: 10': "ranking_2"
    , 'Unnamed: 11': "ranking_3"
    , 'Unnamed: 12': "ranking_4"
    , 'Unnamed: 13': "ranking_5"
    , 'Unnamed: 14': "ranking_6"
}

star_wars = star_wars.rename(columns = new_col_names) 
```

```python
star_wars[star_wars.columns[9:15]].head()
```

## Rankings

Now that we've cleaned up the ranking columns, we can find the highest-ranked movie more quickly.

```python
# Computing the mean of each of the ranking columns
star_wars[star_wars.columns[9:15]].mean()
```

```python
# Visualize with a bar chart

import matplotlib.pyplot as plt
%matplotlib inline

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

The first 3 movies of the Star Wars series have the highest ratings, with the ratings peaking at the 3rd movie. The newer movies were not as big as a hit!

But how many of the surveyed sample actually watched each of these movies?

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

More respondents have watched the newer movies versus the older ones. This is an interesting oberservation, because while ratings were climbing across the first 3 movies, the number of respondents who have watched those 3 were declining.

## Segmenting the data by Star Wars fans

We know which movies the survey population as a whole has ranked the highest. Now let's examine how certain segments of the survey population responded.

```python
# Segment by Star Wars fans vs not fans
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

```python
star_wars_fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True]

# Assume those who did not say "Yes" are not fans
not_star_wars_fans = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] != True]
```

```python
# Star Wars Fans - Most viewed movie and the highest-ranked movie

print("Most viewed by Star Wars fans")
plt.bar(range(6), star_wars_fans[star_wars_fans.columns[3:9]].sum())
```

```python
print("Most highly rated by Star Wars fans")
plt.bar(range(6), star_wars_fans[star_wars_fans.columns[9:15]].mean())
```

High view rates across all movies for Star Wars fans -- which is expected. Numbers climbing for the last couple of movies, perhaps because more fans were converted over the years. However, fans rate the original movies much higher than the newer movies.

```python
# Not Star Wars Fans - Most viewed movie and the highest-ranked movie
print("Most viewed by Non Star Wars fans")
plt.bar(range(6), not_star_wars_fans[not_star_wars_fans.columns[3:9]].sum())
```

```python
print("Most highly rated by non Star Wars fans")
plt.bar(range(6), not_star_wars_fans[not_star_wars_fans.columns[9:15]].mean())
```

Much lower view rates and ratings across all movies from non Star Warn fans.
