## Introduction

In this project, we'll work with data from the CIA World Factbook, a compendium of statistics about all of the countries on Earth. The Factbook contains demographic information like:

- population - The population as of 2015.
- population_growth - The annual population growth rate, as a percentage.
- area - The total land and water area.

In this project, we'll explore the Python SQLite workflow to explore, analyze, and visualize data from this database.

## The database

Let's first try to understand the database and tables within.

```python
import pandas as pd
import sqlite3

# Creating a sqlite3.connection instance
conn = sqlite3.connect("factbook.db")

# Get information on the tables in the database
q = "SELECT * FROM sqlite_master WHERE type = 'table'"
pd.read_sql_query(q, conn)
```

```python
# Returning the first 5 rows of the facts table in the database
q = "SELECT * FROM facts LIMIT 5"
pd.read_sql_query(q, conn)
```

```python
# Returning the column names of the facts table
q = "PRAGMA table_info(facts)"
pd.read_sql_query(q, conn)
```

## Initial exploration - 'facts' table

Let's start by calculating some summary statistics and look for any outlier countries.

```python
# Returning the min and max of population and population_growth
q = "SELECT min(population), max(population), min(population_growth), max(population_growth) from facts"
pd.read_sql_query(q, conn)
```

From the query above, we see that:
- there's a country with a population of 0
- there's a country with a population of 7256490011 (or more than 7.2 billion people)

Let's zoom in on just these countries.

```python
# Returning the countries with 0 population
q = "SELECT id, name, population FROM facts WHERE population = 0"
pd.read_sql_query(q, conn)
```

Antarctica is the only country with 0 population. This makes sense. Let's check the next few countries with lowest population.

```python
# Returning countries with lowest population
q = "SELECT * FROM facts ORDER BY population ASC LIMIT 5"
pd.read_sql_query(q, conn)
```

There are several countries with "None", or no information, on the population. Let's see how many:

```python
# Returning countries with no information on population
q = "SELECT id, name, population FROM facts WHERE population IS NULL"
pd.read_sql_query(q, conn)
```

```python
# Returning count of countries with no information on population
q = "SELECT count(id) FROM facts WHERE population IS NULL"
pd.read_sql_query(q, conn)
```

19 countries with no information on population. Most are small islands.

Let's now take a look at the country with largest population.

```python
# Returning the maximum population
q = "SELECT * FROM facts ORDER BY population DESC LIMIT 1"
pd.read_sql_query(q, conn)
```

The entire world is listed as a separate entry. We should exclude this entry in our analysis.

## Creating histograms

Let's generate histograms for the rest of the countries in the table, ignoring the 'World' and 'Antarica' rows.

```python
# First, let's return all of the values in the columns you want to visualize
q = '''
SELECT name, population, population_growth, birth_rate, death_rate
FROM facts
WHERE name != 'World' AND name != 'Antartica'
'''

data = pd.read_sql_query(q, conn)
data.sort_values(by = ['population'], ascending=False).head()
```

```python
# Visualising the histograms 

import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

data["birth_rate"].hist(ax=ax1)
ax1.set_xlabel("birth_rate")
data["death_rate"].hist(ax=ax2)
ax2.set_xlabel("death_rate")
data["population_growth"].hist(ax=ax3)
ax3.set_xlabel("population_growth")
data["population"].hist(ax=ax4)
ax4.set_xlabel("population")

plt.show()
```

Majority of countries in the world have low birth rates and low population growth -- these are most likely to be developed countries. Interestingly, death rate is more normally distributed than the other indicators. There are several outliers with very large populations, skewing the histogram.
